```python
# Import libraries
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
# Load the train dataset and separate the target
iowa_file_path = "/Users/thitipatchansriniyom/Downloads/DA/Kaggle Competitions/train.csv"
home_data = pd.read_csv(iowa_file_path)
y = home_data["SalePrice"]
```

```python
# Select relevant features
features = ['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
```

```python
# Select columns corresponding to features, and explore these features
X = home_data[features]
X.head()
```

```python
X.tail()
```

```python
X.describe()
```

```python
# Split the train dataset into validation and training data
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=1)
```

```python
# Define a Random Forest model
rf_model = RandomForestRegressor(random_state=1)
rf_model.fit(train_X, train_y)
rf_val_predictions = rf_model.predict(val_X)
rf_val_mae = mean_absolute_error(rf_val_predictions, val_y)

print("Validation MAE for Random Forest Model: {:,.0f}".format(rf_val_mae))
```

```python
# To improve the accuracy of the model, create another Random Forest model and train it on all the training data
rf_model_on_full_data = RandomForestRegressor(random_state=1)
rf_model_on_full_data = rf_model_on_full_data.fit(X, y)
```

```python
# Read the test dataset and make predictions for the submission to Kaggle's Housing Prices competition
test_data_path = "/Users/thitipatchansriniyom/Downloads/DA/Kaggle Competitions/test.csv"
test_data = pd.read_csv(test_data_path)
test_X = test_data[features]
test_preds = rf_model.predict(test_X)
```

```python
# Format the submission file
output = pd.DataFrame({'Id': test_data.Id,
                       'SalePrice': test_preds})
output.to_csv('submission.csv', index=False)
```
